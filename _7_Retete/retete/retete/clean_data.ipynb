```python
#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
title_row = []
content_row = []
ingredients_row = []
recipe_yield = []


def translate_data(row, col):
  translated_value = ''
  translation = translator.translate(
      row,
      src="ro",
      dest='hu'
    ) 
  translated_value = translation.text

  print(translated_value)
  if col == 'Title':
    title_row.append(translated_value)
  elif col == 'Content':
    content_row.append(translated_value)
  elif col == 'Ingredients':
    ingredients_row.append(translated_value)
  elif col == 'Recipe Yield':
    recipe_yield.append(translated_value)


content_table = data.drop(['Page','Url','Image','Portion','Cooking Time'],axis = 1 )
content_table.replace(
        {
            '[\u0370-\u03FF]':'-',
            '\r\n':'-',
              '\xa0': '-'
        },
        regex = True,
        inplace = True
    )
#content_table.loc[77]['Content']
#data['Ingredients']
for col in content_table.columns:
  for index, value in content_table[col].iteritems():
    translate_data(value,col)
```
